```python
# Scatter Matrix
scatter_matrix(data[['calories', 'fat', 'sugars', 'carbo']], figsize=(7, 7))
plt.show()

# Scatter Plot
plt.scatter(data["sugars"], data["calories"], color = "red")
plt.title("Relation between sugars and calories")
plt.xlabel("Sugars")
plt.ylabel("Calories")
plt.show()

plt.scatter(data["sugars"], data["carbo"], color = "green")
plt.title("Relation between sugars and carbs")
plt.xlabel("Sugars")
plt.ylabel("Carbs")
plt.show()

plt.scatter(data["sugars"], data["fat"], color = "blue")
plt.title("Relation between sugars and fat")
plt.xlabel("Sugars")
plt.ylabel("Fat")
plt.show()

plt.scatter(data["carbo"], data["calories"], color = "red")
plt.title("Relation between carbs and calories")
plt.xlabel("Carbs")
plt.ylabel("Calories")
plt.show()

plt.scatter(data["carbo"], data["fat"], color = "green")
plt.title("Relation between carbs and fat")
plt.xlabel("Carbs")
plt.ylabel("Fat")
plt.show()

plt.scatter(data["carbo"], data["sugars"], color = "blue")
plt.title("Relation between carbs and sugars")
plt.xlabel("Carbs")
plt.ylabel("Sugars")
plt.show()

plt.scatter(data["fat"], data["calories"], color = "red")
plt.title("Relation between fat and calories")
plt.xlabel("Fat")
plt.ylabel("Calories")
plt.show()

plt.scatter(data["fat"], data["carbo"], color = "green")
plt.title("Relation between fat and carbs")
plt.xlabel("Fat")
plt.ylabel("Carbs")
plt.show()

plt.scatter(data["fat"], data["sugars"], color = "blue")
plt.title("Relation between fat and sugars")
plt.xlabel("Fat")
plt.ylabel("Sugars")
plt.show()

plt.scatter(data["calories"], data["fat"], color = "red")
plt.title("Relation between fat and calories")
plt.xlabel("Calories")
plt.ylabel("Fat")
plt.show()

plt.scatter(data["calories"], data["carbo"], color = "green")
plt.title("Relation between fat and carbs")
plt.xlabel("Calories")
plt.ylabel("Carbs")
plt.show()

plt.scatter(data["calories"], data["sugars"], color = "blue")
plt.title("Relation between fat and sugars")
plt.xlabel("Calories")
plt.ylabel("Sugars")
plt.show()

```
